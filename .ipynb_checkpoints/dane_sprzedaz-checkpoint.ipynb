```python
import pandas as pd
import os
import matplotlib.pyplot as plt
from itertools import combinations
from collections import Counter

%matplotlib inline 

print("Gotowy do pracy!")
```

```python
path = 'C:/Users/SP/Desktop/jupyter notebook/Sales_Data'
files = [file for file in os.listdir(path) if file.endswith('.csv')]
all_months_data = pd.DataFrame()
```

```python
for file in files:
    current_data = pd.read_csv(os.path.join(path, file))
    all_months_data = pd.concat([all_months_data, current_data], ignore_index=True)
```

```python
print(f"Łączna liczba wierszy: {all_months_data.shape[0]}")
all_months_data.head(10)
```

```python
# czyszczenie wierszy
df = all_months_data.dropna(how='all')
print(f"wiersze po usunieciu NaN: {df.shape[0]}")
df = df[df['Order ID'] != 'Order ID']
print(f"Liczba wierszy bez naglowkow: {df.shape[0]}")
```

```python
# konwersja danych
df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'])
df['Price Each'] = pd.to_numeric(df['Price Each'])
df['Order Date'] = pd.to_datetime(df['Order Date']) 
```

```python
# nowe kolumny do analizy
df['Sales'] = df['Quantity Ordered'] * df['Price Each']
df['Month'] = df['Order Date'].dt.month
df['Hour'] = df['Order Date'].dt.hour
```

```python
# Najlepsze miesiące pod względem sprzedaży, największy ruch jest w okresie świątecznym, zespoły logistyczne oraz magazynowe powinny koncentrować
# się na tym okresie
najlepszy_miesiac = df.groupby('Month').sum('Sales').sort_values(['Sales'], ascending = False)
najlepszy_miesiac
```

```python
def get_city_state(address):
    city = address.split(',')[1].strip()
    state_code = address.split(',')[2].split(' ')[1]
    return f"{city} ({state_code})"

# Tworzenie kolumny City
df['City'] = df['Purchase Address'].apply(lambda x: get_city_state(x))
```

```python
# miasta z największą ilością zamówień
city_orders = df.groupby('City')['Order ID'].count().sort_values(ascending=False)
```

```python
# Najlepsze miejsca na np. wydatki reklamowe (ze względu na ilość wykonywanych zamówień)
cities = city_orders.index
orders = city_orders.values

plt.figure(figsize=(12,6))
plt.bar(cities,orders)
plt.xticks(rotation='vertical', size=8)
plt.title('Liczba Zamówień wg Miasta (Cały Rok)')
plt.xlabel('Miasto (Kod Stanu)')
plt.ylabel('Liczba Zamówień')
```

```python
# zamówienia klientów według godziny, reklamowanie się jest najefektywniejsze w dwóch szczytach

hourly_orders = df.groupby('Hour')['Order ID'].count()
hours = [hour for hour, count in hourly_orders.items()]
order_counts = hourly_orders.values

plt.figure(figsize=(10, 5))
plt.plot(hours, order_counts, marker='o', linestyle='-', color='b')
plt.xticks(hours)
plt.grid()
plt.title('Liczba Zamówień wg Godziny Dnia')
plt.xlabel('Godzina dnia')
plt.ylabel('Liczba Transakcji')
```

```python
# identyfikacja wieloproduktowych zamówień
df_grouped = df[df['Order ID'].duplicated(keep=False)]
df_grouped['Grouped Products'] = df_grouped.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df_grouped = df_grouped[['Order ID', 'Grouped Products']].drop_duplicates()
print(df_grouped.head(10))
```

```python
# liczenie par produktów
count = Counter()
for row in df_grouped['Grouped Products']:
    row_list = row.split(',')

    for combination in combinations(row_list, 2):
        count[combination] += 1
# 10 najczęściej kupowanych par produktów
print(count.most_common(10))
```

```python
# najlepiej sprzedające się produkty pod względem ilościowym, są to produkty tanie oraz pierwszej potrzeby
najpopularniejszy_produkt = df.groupby('Product').sum('Quantity Ordered').sort_values(['Quantity Ordered'], ascending=False)
```
